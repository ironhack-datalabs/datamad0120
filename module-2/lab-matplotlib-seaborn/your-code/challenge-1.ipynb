#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(y)
plt.subplot(1,2,2)
plt.xscale('log')
plt.plot(y)

```

```python
fig, ax1 = plt.subplots()

ax1.plot(y,color="y") # eje x1 -->lineal
ax1.set_xlabel('y=x^2');

ax2 = ax1.twiny() # ax1 y ax2 comparten eje y
ax2.semilogx(y) # eje x2 -->log
ax2.set_xlabel('y=exp(x)');
```
