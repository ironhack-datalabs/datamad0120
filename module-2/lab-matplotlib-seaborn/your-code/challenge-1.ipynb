#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (20, 10)) # crea la fig y le asigna dos ax (ax1, ax2) y crea los subplots
ax1.plot(x, y, '*', color = 'r')
ax2.plot(x, z, '--', color = 'g') # ---> Se le pueden añadir mas ax en la primera línea
ax1.set_xlabel('Tiempo') # Asigna un nombre al eje x del ax 1
ax1.set_ylabel('Ventas') # Asigna un nombre al eje y del ax 1
ax2.set_xlabel('Tiempo') # Asigna un nombre al eje x del ax 2
ax2.set_ylabel('Coches') # Asigna un nombre al eje y del ax 2
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
y = x**2
z = np.exp(x)
plt.figure(figsize = (10, 5)) # crea la figura y le da un tamaño con figsize
plt.subplot(1,2,1) # crea el primer subplot de 1 fila 2 columnas y posición 1
plt.plot(x, y)
plt.plot(x, z)
plt.yscale('log') # asigna escala logaritmica al eje y
plt.title("Logarítmica")
plt.subplot(1,2,2) # crea el primer subplot de 1 fila 2 columnas y posición 2
plt.plot(x, y)
plt.plot(x, z)
plt.title("Normal")
```
