```python
import numpy as np
import pandas as pd
import re
data = pd.read_csv('sharkattack.csv', engine = 'python')
```

```python
#Datos iniciales 5976 filas 23 columnas
#Borrar las columnas que tengan más de 3000 null
null_cols = data.isnull().sum()
null_cols[null_cols > 0]
#Borrar Unnamed:22, Unnamed:23, Time
drop_cols = list(null_cols[null_cols > 3000].index)
data = data.drop(drop_cols, axis=1)              
```

```python
#Borrar las columnas que considero irrelevantes a la hora de analizar los datos
data = data.drop(['Name','Case Number','original order','Investigator or Source', 'pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2'], axis=1)
#Renombrar las columnas para no tener problemas
data = data.rename(columns={'Sex ':'Sex', 'Species ': 'Species'})
```

```python
# Si se quiere analizar lugares no tenemos información de 24 filas (casos)
# en los demás se puede deducir dónde son los accidentes
null_displ = data[(data['Area'].isnull()==True) &  (data['Location'].isnull()==True) 
                  &  (data['Country'].isnull()==True)]
```

```python
#Limpiar datos de Fatal, si es 'N ', ' N', 'n', si es algo irreconocible poner UNKNOWN
data.loc[(data['Fatal (Y/N)']==' N' ) | (data['Fatal (Y/N)']=='N ' ) | (data['Fatal (Y/N)']=='n' ) | (data['Fatal (Y/N)']=='F' ), 'Fatal (Y/N)'] = 'N'
data.loc[(data['Fatal (Y/N)'] != 'N') & (data['Fatal (Y/N)']!='Y') & (data['Fatal (Y/N)']!='UNKNOWN'), 'Fatal (Y/N)'] = 'UNKNOWN'
```

```python
#Dividir la columna fecha para sacar el mes y el año (en caso de ser null año y no null fecha)
false_year = data[data['Year'] == 0]
real_year = false_year['Date']
real_year = real_year.str.extract(r'(\d\d\d\d)')

false_year['Real Year'] = real_year
false_year = false_year.drop('Year', axis = 1)
false_year = false_year.rename(columns={'Real Year':'Year'})

data = data[(data['Year'] != 0)]
data = data.append(false_year, sort = True)
```

```python
column_order = ['Date', 'Year', 'Country', 'Area', 'Location', 'Activity', 'Type',
         'Sex', 'Age', 'Injury', 'Fatal (Y/N)', 'Species',]
data = data[column_order]
```

```python
#Extract month form Date column
s = data['Date']
s = s.str.extract(r'(\d\d)?(-)?(\S\S\S)?(-)?(-)?(\d\d)')
```

```python
data['Month'] = s[2]
```

```python
column_order = ['Year','Month', 'Date', 'Country', 'Area', 'Location', 'Activity', 'Type',
                'Sex', 'Age', 'Injury', 'Fatal (Y/N)', 'Species' ]

data = data[column_order]
```

```python
step = data.groupby(['Date','Country'], as_index = False).size().reset_index().\
            rename(columns={0:'Number of victims'})
data['Number of victims'] = step['Number of victims']
```

```python
data = data.groupby(['Year','Month','Date','Country','Area','Location','Type','Species','Number of victims'], as_index = False).agg({'Age':list,'Injury':list, 'Activity':list,'Fatal (Y/N)': list,
                           'Sex':list})
```

```python
#Clasificar por season
def agg_season(x):
    if x in ['Dec','Jan','Feb']:
        return 'Winter'
    elif x in ['Mar','Apr','May']:
        return 'Spring'
    elif x in ['Jun','Jul','Aug']:
        return 'Summer'
    else:
        return 'Autumn'
   
data['Season'] =  data['Month'].agg(agg_season)
```
