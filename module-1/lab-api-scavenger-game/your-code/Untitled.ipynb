```python
def requestGithub(endpoint):
    token = os.getenv("tokenApi")
    if not token:
        raise ValueError("You must set a tokenApi token")
    
    baseUrl = "https://api.github.com"
    
    url = baseUrl + endpoint
    
    headers = {
        "Authorization": f"token {token}"
    }
    
    res = requests.get(url, headers = headers)
    
    if res.status_code != 200:
        print(res.text)
        raise ValueError("Bad Response")
        
    return res.json()
```

```python
requestGithub("/repos/ironhack-datalabs/datamad0120/forks")
```

# Challenge 3: Hidden Cold Joke

#### Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:

#### https://github.com/ironhack-datalabs/scavenger

#### The filenames of the secret files contain .scavengerhunt and they are scattered in different directories of this repo. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. They are scattered randomly throughout this repo. You need to search for these files by calling the Github API, not searching the local files on your computer.

```python
sha_trees = requestGithub("/repos/ironhack-datalabs/scavenger/branches/master")
sha_number = sha_trees['commit']['sha']
sha_number
```

```python
sha = requestGithub("/repos/ironhack-datalabs/scavenger/git/trees/9308ccc8a4c34c5e3a991ee815222a9691c32476?recursive=1")
sha
```

```python
all_paths = ([p['path'] for p in sha['tree'] if '.scavengerhunt' in p['path']])

print(all_paths)
```

```python
cont = requestGithub(f"/repos/ironhack-datalabs/scavenger/contents/15024/.0006.scavengerhunt")
cont
```

```python
all_contents = []

for p in all_paths:
    all_contents.append((requestGithub(f"/repos/ironhack-datalabs/scavenger/contents/{p}")['name'], requestGithub(f"/repos/ironhack-datalabs/scavenger/contents/{p}")['content']))
    
print(all_contents)
```

```python
sorted_content = sorted(all_contents)
print(sorted_content)
```

```python
final_content = []

for c in sorted_content:
    final_content.append(c[1])

print(final_content)
```

```python
import base64
```

```python
putica_frase = ''

for c in final_content:
    b = base64.b64decode(c)
    b = b.decode("utf-8")
    putica_frase += b.replace("\n"," ")
    
putica_frase.strip()
    
```
