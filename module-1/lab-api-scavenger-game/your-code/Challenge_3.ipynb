```python
def requestGithub(endpoint):
    token = os.getenv("githtok")
    if not token:
        raise ValueError("You must set a GITHUB_APIKEY token")
    
    baseUrl = "https://api.github.com"
    url = baseUrl+endpoint
    print(f"Pidiendo movidas a  {url}")
    headers = {
        "Authorization": f"token {token}"
    }
    res = requests.get(url,headers=headers)
    if res.status_code != 200:
        print(res.text)
        raise ValueError("Bad Response")
    return res.json()
```

```python
contents = requestGithub("/repos/ironhack-datalabs/scavenger/contents?extension:scavengerhunts")
print("Hay ", len(contents) ," archivos con la extensión .scavengerhunts")
```

```python
def request(x):
    token = os.getenv("githtok")
    if not token:
        raise ValueError("You must set a GITHUB_APIKEY token")
    
    headers = {
        "Authorization": f"token {token}"
    }
    
    url = f'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{x}'
    res = requests.get(url,headers=headers)
    print(f"Solicitando data de {url}")
    
    if res.status_code != 200:
        print(res.text)
        raise ValueError("Bad Response")
    return res.json()
    
```

```python
contenido = list(map(request,carpetas))
```

```python
contenido
```

```python
# me quedo con los valores de path cuyo nombre contenga ".scavengerhunt"
```

```python
lista_paths = [y["path"] for x in contenido for y in x]
```

```python
files = [a for a in lista_paths if ".scavengerhunt" in a]
```

```python
print(files)
```

### ordeno según el índice 9 y 10 de cada string.

```python
ordenado = sorted(files, key=lambda x: (x[9],x[10]))
```

### hago la request de todos los archivos ordenados:

```python
archivos = list(map(request,ordenado))
```

```python
contenido = [x["content"] for x in archivos]
```

### decodifico

```python
import base64
decod = [base64.b64decode(b) for b in contenido]
print(decod)
```

```python
final = [i.decode("UTF-8") for i in decod]
print(final)
```

```python
texto = [z.replace("\n"," ")for z in final]
```

```python
FIN = "".join(texto)
print(FIN)
```
